# Filter Data

```python
lmark_data=pd.read_csv('drive/Landmark_Recognition/lmark_data.csv')
```

```python
lmark_data.head()
```

```python
lmark_data_filtered=pd.DataFrame()
```

```python
count=0
total=0
for i in lmark_data.landmark_id.unique()[0:248]:
  lid=len(lmark_data.loc[lmark_data['landmark_id']==i])
  if not(lid>300) and not(lid<5):
    lmark_data_filtered=lmark_data_filtered.append(lmark_data.loc[lmark_data['landmark_id']==i], ignore_index=True)
    count+=1
    print(lid)
    total+=lid
    
```

```python
lmark_data_filtered.tail()
```

```python
lmark_data_filtered.nunique()
```

```python
lmark_data_filtered.to_csv('drive/Landmark_Recognition/lmark_data_200.csv')
```
